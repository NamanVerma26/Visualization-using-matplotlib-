```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("cleaned.csv")
```

```python
Item_Count=df['Item'].value_counts().sort_values(ascending=False)
Item_Count
```

```python
plt.bar(Item_Count.index, Item_Count)
plt.xlabel('Item')
plt.ylabel('Count')
plt.title('Items Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
p=df['Payment Method'].value_counts()
l=df['Payment Method'].unique().tolist()
plt.pie(p, labels=l, startangle=90, autopct="%1.1f%%")
# plt.legend()
plt.show()
```

```python
fig, ax= plt.subplots(1, 2)
ax[0].pie(Item_Count, labels=df['Item'].unique().tolist(), autopct="%1.1f%%")
ax[1].pie(p, labels=l, autopct="%1.1f%%")

fig.tight_layout()

plt.show()
```
